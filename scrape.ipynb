```python
import re
def scrape_one_page(page_url):
    soup = BeautifulSoup(requests.get(page_url).content)
    page_captions = soup.find("div", attrs={"class":"descContainer"}).find_all("div", attrs={"class":"captionBottom"})

    names = [cap.a.text.strip() for cap in page_captions]
    height_pattern = r"\d+ft\s?\d{1,2}|\d{1,2}ft"
    heights = [re.findall(height_pattern, cap.text)[0] for cap in page_captions]

    return dict(zip(names, heights))
```

```python
LETTERS = [l.upper() for l in "a b c d e f g h i j k l m n o p q r s t u v w x y z".split()]
```

```python
height_dict = {}
for letter in LETTERS:
    print(f"Scraping {letter}")
    url = f"https://www.celebheights.com/s/{letter}.html"
    soup = BeautifulSoup(requests.get(url).content)
    try:
        max_pages = int(soup.find("ul", attrs={"class":"pagination"}).find_all("li")[-2].get_text())
    except Exception as e:
        print(e)
        continue
    for i in range(1, max_pages):
        if i == 1:
            print("Page 1")
            page_url = url
        else:
            print(f"Page {i}")
            page_url = f"https://www.celebheights.com/s/{letter}{i}.html"
        page_dict = scrape_one_page(page_url)
        height_dict.update(page_dict)
```

```python
len(height_dict)
```

```python
import pandas as pd
heights = pd.DataFrame.from_dict(height_dict, orient="index", columns=["height"]).reset_index().rename(columns={"index":"name"})
heights
```

```python
import pandas as pd

chunk_size = 10000  # Adjust based on your memory
chunks = []

for chunk in pd.read_csv('name.basics.tsv', sep='\t', chunksize=chunk_size):
    # Filter for actors only
    actors_chunk = chunk[chunk['primaryProfession'].str.contains('actor|actress', na=False)]
    chunks.append(actors_chunk)

# Combine all actor chunks
actors_df = pd.concat(chunks, ignore_index=True)
```

```python
living_actors = actors_df[(actors_df.deathYear == "\\N") & (actors_df.birthYear != "\\N")].reset_index()
```

```python
living_actors
```

```python
df = living_actors.merge(heights, left_on="primaryName", right_on="name")[["name", "birthYear", "height"]]
df
```
